```python
import pandas as pd
import operator
import plotly.express as px
%matplotlib inline

from scipy import interpolate
```

# Steps that need to be followed:

1. Get the covid cases data
2. Plot the cases of any 3 countries
3. Plot the vaccination rate of these countries

```python
path='https://covid.ourworldindata.org/data/owid-covid-data.csv'
all_countries=pd.read_csv(path,sep=',')
```

```python
all_countries.to_csv('covid_new_data_covid.csv',';')
```

```python
data_italy=all_countries[all_countries['location']=='Italy']
```

```python
data_italy = data_italy.reset_index(drop=True)
```

```python
#data_italy
```

# TASK 1

The relative cases of Covid-19 over time for one country:

```python
data_italy['total_cases_relative'] = data_italy['total_cases']/data_italy['population']
```

```python
fig = px.line(data_italy, x="date", y="total_cases", title='Covid cases in Italy')
fig.show()
```

The relative cases of Covid-19 for 3 countries:

```python
data_canada = all_countries['location'] == 'Canada'
```

```python
data_spain = all_countries['location'] == 'Spain'
```

```python
data_italy = all_countries['location'] == 'Italy'
```

```python
data_all = data_italy|data_canada|data_spain
```

```python
data_all_countries = all_countries[data_all]
```

```python
data_all_countries=data_all_countries.reset_index(drop=True)
```

```python
#data_all_countries.columns
```

```python
fig = px.line(data_all_countries, x="date", y="total_cases", color='location', title='Covid cases in 3 countries')
fig.show()
```

```python
data_all_countries['total_cases_relative']=data_all_countries['total_cases']/data_all_countries['population']
```

```python
fig = px.line(data_all_countries, x="date", y="total_cases_relative", color='location', title='Relative Covid cases in all 3 countries')
fig.show()
```

# TASK 2

The vaccination rate of the three countries over time: (i.e., % of the population vaccinated)

```python
data_all_countries['vaccination_rate']=(data_all_countries['people_fully_vaccinated']/data_all_countries['population'])*100
```

```python
fig = px.line(data_all_countries, x="date", y="vaccination_rate", color='location', title='Vaccination rate in 3 countries over time')
fig.show()
```

## Vaccination stats with interpolated data

```python
data_all_countries['people_fully_vaccinated_interpolated'] = data_all_countries['people_fully_vaccinated'].interpolate(method="pad", limit=2)

#data_all_countries.columns
data_all_countries['people_fully_vaccinated_interpolated'].interpolate(method="pad", limit=2)
```

```python
data_all_countries['people_fully_vaccinated_new']=(data_all_countries['people_fully_vaccinated_interpolated']/data_all_countries['population'])
```

```python
fig = px.line(data_all_countries, x='date', y='people_fully_vaccinated_new', color='location', title='Total vaccinations in all 3 countries')
fig.show()
```
